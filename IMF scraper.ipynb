```python
# %%
import pandas as pd 
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
import time

# %%
options = webdriver.ChromeOptions()
driver = webdriver.Chrome('/Users/husnainmustafa/Documents/Thesis/chromedriver', options=options)

# Navigate to the IMF news search page
driver.get('https://www.imf.org/en/news/searchnews#sort=%40imfdate%20descending')
NTYPE = Select(driver.find_element(By.ID, 'NewsType'))
NTYPE.select_by_value('Communique')
driver.find_element(By.CLASS_NAME, 'coveo-custom-filter-button').click()

# Initialize an empty list to store links
all_links = []

# Loop through all pages
while True:
    # Wait for the links to load
    time.sleep(2)  # Adjust sleep time as necessary
    
    # Try to fetch the links
    try:
        links = driver.find_elements(By.CSS_SELECTOR, 'a.CoveoResultLink')
        
        # Store the links in the list
        for link in links:
            all_links.append(link.get_attribute('href'))
        
        # Check for the next page button
        next_button = driver.find_element(By.CLASS_NAME, 'coveo-pager-next')
        next_button.click()  # Click the next button
        
    except Exception as e:
        print(f"An error occurred: {e}")
        print("No more pages to navigate or stale element reference.")
        break  # Exit the loop if an error occurs

# Print all collected links
for link in all_links:
    print(link)

# Close the driver
driver.quit()
```
